```python
path_fake ="/content/Fake.csv"
path_true ="/content/True.csv"

# Read both files
df_fake = pd.read_csv(path_fake, delimiter=",",
                 error_bad_lines=False,
                 encoding='utf-8')
df_true = pd.read_csv(path_true,  delimiter=",",
                 error_bad_lines=False,
                 encoding='utf-8')

# Set value  0 to fake news and value 1 to true news
df_fake['label'] = 0
df_true['label'] = 1
```

```python
df = pd.concat([df_fake[0:100], df_true[0:100]])
```

```python
df
```

```python
df.columns
```

```python
df.label.value_counts()
```

```python
data = df[['text', 'label']]
```

```python
y = list(data['label'])
```

```python
# Word Vectorizing usinf TFIDF


documents=[]

for i in data['text']:
    s = '''{}'''.format(str(i))
    documents.append(s)
vectorizer = TfidfVectorizer(stop_words='english')
x = vectorizer.fit_transform(documents)
```

```python
documents[0]
```

```python
x[0]
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.15,
                                                    random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(x_train, y_train)
```

```python
lr_clf.score(x_test, y_test)
```

```python
# Word Embeddings using BERT

documents_bert=[]

for i in data['text']:
    s = '''{}'''.format(str(i))
    documents_bert.append(s)

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens') #bert-base-nli-mean-tokens
embeddings = model.encode(documents_bert)
print(embeddings)
```

```python
embeddings.shape
```

```python

train_features, test_features, train_labels, test_labels = train_test_split(embeddings, y, test_size = 0.10)
```

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression()
lr_clf.fit(train_features, train_labels)
```

```python
lr_clf.score(test_features, test_labels)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

rf_clf = RandomForestClassifier(max_depth=2, random_state=0)
rf_clf.fit(train_features, train_labels)
```

```python
rf_clf.score(test_features, test_labels)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

documents=[]

for i in data['text']:
    s = '''{}'''.format(str(i))
    documents.append(s)

import nltk
# from nltk.tokenize import sent_tokenize
from nltk.stem import PorterStemmer
new_doc = []
for sentence in documents[0:5]:
  sentence = sentence.lower()

  word_tokens = nltk.word_tokenize(sentence)
  print(word_tokens)

  stemmer = PorterStemmer()
  stemmed_word = []

  for token in word_tokens:
      stemmed_word.append(stemmer.stem(token))


  new_doc.append(" ".join(stemmed_word))

```
